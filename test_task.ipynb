```python
import json

with open("input.json") as jsonFile:
    jsonObject = json.load(jsonFile)
    jsonFile.close()

amount_credit = jsonObject['amount_credit']
monthly_pay = jsonObject['monthly_pay']
interest_rate = jsonObject['interest_rate']

period = jsonObject['period'] # number of months


def calculate_interest_paid(amount_credit,
                            monthly_pay,
                            interest_rate,
                            period):
    month_percent = interest_rate / 12
    body_credit = amount_credit
    interest_paid = 0

    for _ in range(period):
        if body_credit >= monthly_pay:
            interest_paid += body_credit * (month_percent / 100)
            body_credit -=  (monthly_pay - body_credit * (month_percent / 100))
        else:
            interest_paid += body_credit * (month_percent / 100)
            body_credit = 0
            break
    
    return interest_paid 

print(f'Interest paid for {period} months : \
{round(calculate_interest_paid(amount_credit, monthly_pay,interest_rate, period),2)}')
```
